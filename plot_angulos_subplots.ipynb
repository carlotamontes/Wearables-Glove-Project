# Plot CSV files as subplots and save PNG

This notebook reads the 7 CSV files inside the `Angulos` folder, plots them as subplots in one figure, and saves the figure as a PNG.

Expected structure:

```text
Wearables-Glove-Project/
├── Angulos/
│   ├── 50kohm-1.csv
│   ├── 50kohm-2.csv
│   ├── 50kohm-3.csv
│   ├── 150kohm.csv
│   ├── 210kohm.csv
│   ├── 250kohm.csv
│   └── 460kohm.csv
└── plot_angulos_subplots.ipynb
```

````python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import math

# Folder containing your CSV files
folder = Path("Angulos")

files = [
    "50kohm-1.csv",
    "50kohm-2.csv",
    "50kohm-3.csv",
    "150kohm.csv",
    "210kohm.csv",
    "250kohm.csv",
    "460kohm.csv",
]

# Output image name
output_png = "angulos_subplots.png"
````

````python
# Create subplot grid
n_files = len(files)
n_cols = 2
n_rows = math.ceil(n_files / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 3.2 * n_rows), sharex=False)
axes = axes.flatten()

for ax, file_name in zip(axes, files):
    file_path = folder / file_name
    
    df = pd.read_csv(file_path)
    
    # Assumes column 1 = time, column 2 = values
    time = df.iloc[:, 0]
    values = df.iloc[:, 1]
    
    ax.plot(time, values)
    ax.set_title(file_name.replace(".csv", ""))
    ax.set_xlabel("Time")
    ax.set_ylabel("Value")
    ax.grid(True)

# Hide unused subplot, because 7 files in a 4x2 grid leaves 1 empty axis
for ax in axes[n_files:]:
    ax.set_visible(False)

fig.suptitle("Angle calibration CSV plots", fontsize=16)
fig.tight_layout()

# Save as PNG in the same folder as the notebook/script
fig.savefig(output_png, dpi=300, bbox_inches="tight")

plt.show()

print(f"Saved PNG: {Path(output_png).resolve()}")
````

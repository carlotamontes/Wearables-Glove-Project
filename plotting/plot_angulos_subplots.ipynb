Expected structure:

```text
Wearables-Glove-Project/
├── Angulos/
│   ├── 50kohm-1.csv
│   ├── 50kohm-2.csv
│   ├── 50kohm-3.csv
│   ├── 150kohm.csv
│   ├── 210kohm.csv
│   ├── 250kohm.csv
│   └── 460kohm.csv
└── plot_angulos_subplots.ipynb
```
